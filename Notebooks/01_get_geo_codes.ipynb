```python
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
def get_lat_long(address):
    geolocator = Nominatim(user_agent="Get_Geo_Codes")
    location = geolocator.geocode(address)
    if location is not None:
        lat_ =location.latitude
        long_ = location.longitude
    else:
        lat_ =long_ = None
        
    return lat_,long_
    
```

```python
all_df = pd.read_csv('../data/combine_3_cities.csv')
```

```python
lat =[]
long=[]
for index,data in all_df.iterrows():
    address = f'{data["Neighbourhood"]} {data["City"]} {data["Country"]}'
    lat_ , long_ =get_lat_long(address)
    lat.append(lat_)
    long.append(long_)
```

```python
all_df['Latitude']=lat
all_df['Longitude']=long
```

## Getting some of the missing *values*

```python
for index , data in all_df[(all_df.Latitude.isna())].iterrows():
    address = f'{data["Neighbourhood"]},{data["Country"]}'
    lat_ , long_ =get_lat_long(address)
    all_df.loc[all_df['Neighbourhood']==data['Neighbourhood'],'Latitude']=lat_
    all_df.loc[all_df['Neighbourhood']==data['Neighbourhood'],'Longitude']=long_
    print('The geograpical coordinate of {} are {}, {}.'.format(address, lat_, long_))
```
